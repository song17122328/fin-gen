# 阶段一，提取Embedding并可视化

## 步骤一，提取 Embeddings --

```python
import torch
import numpy as np 
import pandas as pd
from pytorch_tabular import TabularModel # 确保 TabularModel 被导入
from torch.utils.data import DataLoader
import os

#避免线程爆炸，线程爆炸会导致kernel重启
os.environ['OPENBLAS_NUM_THREADS'] = '64'
os.environ['MKL_NUM_THREADS'] = '64' # MKL是另一个数学库，以防万一也设置一下
os.environ['OMP_NUM_THREADS'] = '64' # OpenMP是另一个并行计算标准，也设置一下
```

```python
print("-----开始加载数据-----")
import kagglehub
import os
dataset_dir = kagglehub.dataset_download("uciml/default-of-credit-card-clients-dataset")
dataset_path = os.path.join(dataset_dir,"UCI_Credit_Card.csv")
df = pd.read_csv(dataset_path)
df = df.drop("ID",axis = 1)
df = df.rename(columns={'default.payment.next.month': 'default', 'PAY_0': 'PAY_1'})

print("======数据加载完成，开始加载模型=====")
load_path = "saved_models/ft_transformer_v1"
loaded_model = TabularModel.load_model(load_path)

print(f"\n --- 模型从 {load_path} 加载成功！，开始提取整个数据集的Embeddings -----")
# 1. 获取已经配置好的 datamodule
#    这个对象知道如何将你的 DataFrame 转换成模型需要的 Tensors
datamodule = loaded_model.datamodule

# 2. 创建一个用于推理(Inference)的数据加载器(DataLoader)
#    这个加载器会处理所有的数据预处理步骤
inference_dataloader = datamodule.prepare_inference_dataloader(df)

# 3.将底层设置为评估模式
model = loaded_model.model
model.eval()
# 确保模型在 GPU 上（如果可用）
if torch.cuda.is_available():
    model.cuda()

#4.迭代数据加载器，手动提取Embeddings
all_embeddings = []
ifprintshape = False
with torch.no_grad(): #推理关闭梯度计算
    for batch in inference_dataloader:
        #将数据批次移动到与模型相同的设备(CPU或者GPU)
        for k,v in batch.items():
            batch[k] = v.to(model.device)

        #直接调用底层模型的embedding_layer来获取嵌入
        #FT-Transformer 的 embedding_layer接受整个批次作为输入
        batch_embeddings_3d = model.embedding_layer(batch) # 这是一个3D张量

        # 将3D张量 (batch_size, num_features, embedding_dim) 压扁成2D张量
        # 我们使用 .reshape() 或 .view() 来实现
        
        batch_size = batch_embeddings_3d.shape[0]
        batch_embeddings_2d = batch_embeddings_3d.reshape(batch_size, -1)

        if ifprintshape is False:
            ifprintshape = True
            print(f"三维Embedding的形状为:{batch_embeddings_3d.shape}")
            print(f"二维Embedding的形状为:{batch_embeddings_2d.shape}")
            
        #将提取出的embedding从GPU移动到CPU并转为Numpy数组
        all_embeddings.append(batch_embeddings_2d.detach().cpu().numpy())

#5. 将所有批次的embeddings合并成一个大的Numpy数组
all_embeddings = np.concatenate(all_embeddings,axis = 0)

print(f"成功提取了 {all_embeddings.shape[0]} 条数据的 Embeddings。")
print(f"每个 Embedding 的维度是: {all_embeddings.shape[1]}")
print(f"Embeddings 的数据类型: {type(all_embeddings)}")

# 保存 embeddings 以备后用
np.save("all_embeddings.npy", all_embeddings)
print("Embeddings 已保存到 all_embeddings.npy 文件中。")
```

## 步骤二，可视化 Embeddings

```python
import numpy as np
import pandas as pd
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns

print("----开始加载保存的Embeddings----")
all_embeddins = np.load("all_embeddings.npy")

# (我们也需要原始数据中的 'default' 标签来进行上色，所以重新加载一下df)
import kagglehub
import os
dataset_dir = kagglehub.dataset_download("uciml/default-of-credit-card-clients-dataset")
dataset_path = os.path.join(dataset_dir,"UCI_Credit_Card.csv")
df = pd.read_csv(dataset_path)
df = df.drop("ID",axis = 1)
df = df.rename(columns={'default.payment.next.month': 'default', 'PAY_0': 'PAY_1'})
labels = df['default'].values

num_samples_to_plot = 30000 # 先从10000个样本开始，如果还死机，可以降到5000
if len(all_embeddings) > num_samples_to_plot:
    print(f"--- 数据量较大，随机抽取 {num_samples_to_plot} 个样本进行可视化 ---")
    # 生成随机索引
    random_indices = np.random.choice(len(all_embeddings), num_samples_to_plot, replace=False)
    # 根据索引选择对应的 embeddings 和 labels
    embeddings_subset = all_embeddings[random_indices]
    labels_subset = labels[random_indices]
else:
    embeddings_subset = all_embeddings
    labels_subset = labels
print(f"--- 开始使用 t-SNE 对 {len(embeddings_subset)} 个样本进行降维 (这可能需要几分钟) ---")# 初始化 t-SNE 模型
# n_components=2 表示我们想降到二维
# perplexity 是一个关键参数，通常在5-50之间，30是一个很好的起点
# random_state 保证每次运行结果一致
tsne = TSNE(n_components=2, perplexity=30, random_state=42, max_iter=1000)
embeddings_2d = tsne.fit_transform(embeddings_subset)
print("--- 降维完成，开始绘图 ---")

# 创建一个 DataFrame 以便使用 seaborn 绘图
plot_df = pd.DataFrame(embeddings_2d, columns=['tsne_1', 'tsne_2'])
plot_df['default'] = labels_subset
plot_df['default'] = plot_df['default'].map({0: 'No Default', 1: 'Default'}) # 将0/1标签换成更易读的文字

# 使用 seaborn 绘制散点图
plt.figure(figsize=(12, 10))
sns.scatterplot(
    x="tsne_1", y="tsne_2",
    hue="default",
    palette=sns.color_palette("hls", 2), # 使用hls调色盘，指定两种颜色
    data=plot_df,
    legend="full",
    alpha=0.6 # 设置点的透明度，方便观察重叠区域
)

plt.title('t-SNE Visualization of Customer Embeddings', fontsize=16)
plt.xlabel('t-SNE Dimension 1', fontsize=12)
plt.ylabel('t-SNE Dimension 2', fontsize=12)
plt.legend(title='Customer Status')

output_filename = "tsne_visualization.png"
plt.savefig(output_filename, dpi=300, bbox_inches='tight') # dpi提高图片清晰度, bbox_inches='tight'确保所有元素都被保存
print(f"图像已成功保存到: {output_filename}")

# 2. 在屏幕上显示图像
plt.show()

# (可选) 清理画布，为下一次绘图做准备
plt.close()
```

## 步骤三，使用DBSCAN识别稀疏区域

```python
# --- 第三周任务：步骤三，使用DBSCAN识别稀疏/风险区域 ---

import numpy as np
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

print("--- 开始加载完整的 Embeddings 进行聚类 ---")
# 我们在完整的 30000 个样本上进行聚类，以获得最准确的密度视图
all_embeddings = np.load("all_embeddings.npy")

# (我们也需要原始数据中的 'default' 标签，如果不在内存中，需要重新加载)
print("--- 正在加载原始数据以获取标签 ---")
import kagglehub
import os
dataset_dir = kagglehub.dataset_download("uciml/default-of-credit-card-clients-dataset")
dataset_path = os.path.join(dataset_dir,"UCI_Credit_Card.csv")
df = pd.read_csv(dataset_path)
df = df.drop("ID",axis = 1)
df = df.rename(columns={'default.payment.next.month': 'default', 'PAY_0': 'PAY_1'})


print("--- 对 Embeddings 进行标准化处理 ---")
# DBSCAN 对特征的尺度很敏感（即距离的计算），所以在聚类前进行标准化是一个非常好的习惯
scaler = StandardScaler()
embeddings_scaled = scaler.fit_transform(all_embeddings)

print("--- 开始运行 DBSCAN 聚类 (这可能需要一些时间) ---")
# DBSCAN 的两个关键参数:
# eps: 一个点周围多大范围内可以被看作是“邻居”。这个值需要根据你的数据分布来调整。
#      可以从一个较小的值开始尝试。
# min_samples: 一个核心点需要有多少个邻居才能形成一个簇。
# n_jobs=-1 使用所有可用CPU核心加速（它会遵守我们设置的环境变量限制）
dbscan = DBSCAN(eps=3.0, min_samples=15, n_jobs=-1)

# 在标准化后的数据上进行聚类
clusters = dbscan.fit_predict(embeddings_scaled)

print("--- 聚类完成，分析结果 ---")
# DBSCAN 会将噪声点（即稀疏区域的点）的标签设置为 -1
# 让我们看看找到了多少个簇，以及多少个噪声点

# 将聚类结果添加到我们的 DataFrame 中，以便分析
df['cluster_id'] = clusters

num_clusters = len(set(clusters)) - (1 if -1 in clusters else 0)
num_noise_points = np.sum(clusters == -1)

print(f"\nDBSCAN 找到了 {num_clusters} 个簇。")
print(f"识别出了 {num_noise_points} 个噪声点 (位于稀疏区域)。")

# --- 结果分析与验证 ---
print("\n--- 各簇的违约率分析 ---")
# -1 代表稀疏区域的噪声点
print("\n稀疏区域 (噪声点/离群点, Cluster ID: -1) 的统计:")
print(df[df['cluster_id'] == -1]['default'].value_counts(normalize=True))

# 查看最大的几个簇的情况
print("\n主要客户群体的统计 (按簇大小排序):")
print(df['cluster_id'].value_counts().head())

# 我们可以检查一下最大的簇（通常是Cluster 0）的违约情况
if 0 in df['cluster_id'].value_counts():
    print("\n最大客户群体 (Cluster 0) 的统计:")
    print(df[df['cluster_id'] == 0]['default'].value_counts(normalize=True))
```

# 阶段二 筛选用户画像

## 步骤一，构建客户簇画像

```python
import pandas as pd
import numpy as np

profile_features = [
    'LIMIT_BAL', 'SEX', 'EDUCATION', 'MARRIAGE', 'AGE',
    'PAY_1', 'BILL_AMT1', 'PAY_AMT1'
]

# 我们关注数值特征的均值，和类别特征的众数（最常见的值）
agg_functions = {
    'LIMIT_BAL': 'mean',
    'AGE': 'mean',
    'BILL_AMT1': 'mean',
    'PAY_AMT1': 'mean',
    'SEX': lambda x: x.mode()[0],
    'EDUCATION': lambda x: x.mode()[0],
    'MARRIAGE': lambda x: x.mode()[0],
    'PAY_1': lambda x: x.mode()[0],
    'default': ['mean', 'count'] # 'mean' of a 0/1 column is the rate, 'count' is the size
}

print("--- 正在计算每个簇的统计画像 ---")
# 使用 groupby 对每个 cluster_id 进行聚合统计
cluster_profiles = df.groupby('cluster_id').agg(agg_functions)

# 为了方便阅读，稍微重命名一下列
cluster_profiles.columns = ['_'.join(col).strip() for col in cluster_profiles.columns.values]
cluster_profiles = cluster_profiles.rename(columns={'default_mean': 'default_rate', 'default_count': 'cluster_size'})

# 按照簇的大小（人数）进行排序
cluster_profiles = cluster_profiles.sort_values(by='cluster_size', ascending=False)


print("\n--- 各客户群体的画像（部分展示） ---")
# 使用 .style.background_gradient() 让表格更易读，高亮显示数值大小
display(cluster_profiles.head(15).style.background_gradient(cmap='viridis'))

# 单独查看一下稀疏区域 (-1) 的画像
print("\n--- 稀疏/噪声群体 (Cluster -1) 的详细画像 ---")
display(cluster_profiles.loc[-1])
```

## 步骤二，筛选高价值目标

```python
# --- 第二阶段：步骤二，筛选高价值目标 ---

print("\n--- 筛选目标客户群体 ---")

# 定义筛选条件
high_income_threshold = df['LIMIT_BAL'].quantile(0.75) # 信用额度在前25%的定义为“高收入”
high_default_rate_threshold = 0.50 # 违约率超过50%的定义为“高风险簇”

low_income_threshold = df['LIMIT_BAL'].quantile(0.25) # 信用额度在后25%的定义为“低收入”
low_default_rate_threshold = 0.10 # 违约率低于10%的定义为“低风险簇”


# 筛选1：高收入 + 高风险的簇
high_risk_vip_clusters = cluster_profiles[
    (cluster_profiles['LIMIT_BAL_mean'] >= high_income_threshold) &
    (cluster_profiles['default_rate'] >= high_default_rate_threshold)
]

print("\n--- 目标1：高收入、高风险的罕见群体 ---")
if not high_risk_vip_clusters.empty:
    display(high_risk_vip_clusters)
else:
    print("未找到完全符合条件的'高收入高风险'簇，可能需要放宽阈值。")
    # 尝试只看高收入群体中，风险相对最高的
    high_income_clusters = cluster_profiles[cluster_profiles['LIMIT_BAL_mean'] >= high_income_threshold]
    if not high_income_clusters.empty:
        print("高收入群体中，风险最高的几个是：")
        display(high_income_clusters.sort_values(by='default_rate', ascending=False).head())


# 筛选2：低收入 + 低风险（坚韧）的簇
resilient_clusters = cluster_profiles[
    (cluster_profiles['LIMIT_BAL_mean'] <= low_income_threshold) &
    (cluster_profiles['default_rate'] <= low_default_rate_threshold)
]

print("\n--- 目标2：低收入、但表现坚韧的群体 ---")
if not resilient_clusters.empty:
    display(resilient_clusters)
else:
    print("未找到完全符合条件的'低收入低风险'簇，可能需要调整阈值。")
```

## 步骤三，为生成器设定中心

```python
# --- 第二阶段：步骤三，为生成器设定靶心 ---

# 假设我们从上一步中，选择了一个高风险VIP簇和一个坚韧簇作为目标
# (请根据你上一步的输出，替换下面的 cluster_id)
# 例如，我们假设发现 42号簇 是高风险VIP, 15号簇 是坚韧群体
target_high_risk_id = high_risk_vip_clusters.index[0] if not high_risk_vip_clusters.empty else -1 # 示例
target_resilient_id = resilient_clusters.index[0] if not resilient_clusters.empty else -1 # 示例

print(f"\n选择的目标 -> 高风险VIP簇ID: {target_high_risk_id}, 坚韧簇ID: {target_resilient_id}")

# 计算这些目标簇在高维 embedding 空间中的中心点（质心）
def get_cluster_centroid(cluster_id):
    # 筛选出属于该簇的所有样本的 embeddings
    cluster_embeddings = all_embeddings[df['cluster_id'] == cluster_id]
    if len(cluster_embeddings) == 0:
        return None
    # 计算均值，即质心
    centroid = np.mean(cluster_embeddings, axis=0)
    return centroid

# 获取质心
centroid_high_risk_vip = get_cluster_centroid(target_high_risk_id)
centroid_resilient = get_cluster_centroid(target_resilient_id)
# 我们也需要一个“普通”客户的起点，就用最大的簇(Cluster 0)的质心
centroid_normal = get_cluster_centroid(0)


# 保存这些关键的“靶心”向量，为第三阶段做准备
if centroid_high_risk_vip is not None:
    np.save("centroid_high_risk_vip.npy", centroid_high_risk_vip)
    print(f"\n高风险VIP簇的质心已保存到 'centroid_high_risk_vip.npy'")
    print(f"向量维度: {centroid_high_risk_vip.shape}")

if centroid_resilient is not None:
    np.save("centroid_resilient.npy", centroid_resilient)
    print(f"坚韧簇的质心已保存到 'centroid_resilient.npy'")
    
if centroid_normal is not None:
    np.save("centroid_normal.npy", centroid_normal)  
    print(f"普通客户簇的质心已保存到 'centroid_normal.npy'")
```

# 第三阶段 生成虚假数据

## 步骤一 准备模型、损失(奖励)函数等

```python
import numpy as np
import pandas as pd
import torch
from pytorch_tabular import TabularModel

# --- 第三阶段：步骤一，准备生成器所需的一切 ---

print("--- 1. 加载所有必需的组件 ---")

# a. 加载我们训练好的、能将数据转换成Embedding的模型
print("加载模型...")
loaded_model = TabularModel.load_model("saved_models/ft_transformer_v1")
model = loaded_model.model
model.eval()
if torch.cuda.is_available():
    model.cuda()

# b. 加载我们选择的起点和终点“靶心”
print("加载目标质心...")
centroid_normal = np.load("centroid_normal.npy")
centroid_high_risk_vip = np.load("centroid_high_risk_vip.npy")

# c. 准备一个“普通客户”作为我们的出发点
# 我们从最大的簇(Cluster 0)中随机选择一个真实的、未违约的客户
df_normal_customers = df[(df['cluster_id'] == 0) & (df['default'] == 0)]
start_customer_series = df_normal_customers.sample(1, random_state=42).iloc[0]

# 将Series转换成模型需要的DataFrame格式（单行）
start_customer_df = pd.DataFrame(start_customer_series).T

print("\n--- 2. 定义我们的“导航”工具 ---")

# a. 定义一个函数，用于将DataFrame数据转换成Embedding
def get_embedding_for_df(input_df):
    datamodule = loaded_model.datamodule
    inference_dataloader = datamodule.prepare_inference_dataloader(input_df)
    with torch.no_grad():
        for batch in inference_dataloader:
            for k, v in batch.items():
                batch[k] = v.to(model.device)
            
            batch_embeddings_3d = model.embedding_layer(batch)
            batch_size = batch_embeddings_3d.shape[0]
            batch_embeddings_2d = batch_embeddings_3d.reshape(batch_size, -1)
            
            return batch_embeddings_2d.detach().cpu().numpy()[0] # 返回单行的embedding

# b. 定义我们的“罗盘”——奖励函数
def calculate_reward(current_embedding, target_embedding):
    """计算当前embedding与目标的距离，距离越小，奖励越高"""
    distance = np.linalg.norm(current_embedding - target_embedding)
    return -distance # 我们用负距离作为奖励



modifiable_features = ['LIMIT_BAL', 'AGE', 'BILL_AMT1', 'PAY_AMT1']

# c. 定义我们的“微调”操作——修改特征的函数
def modify_feature(customer_df):
    """随机选择一个特征并进行微小的、合乎逻辑的修改"""
    # 复制一份以避免修改原始数据
    new_customer_df = customer_df.copy()
    
    # 随机选择一个可修改的特征
    # 我们先从几个简单的数值特征开始
    feature_to_modify = np.random.choice(modifiable_features)
    
    # 定义修改的逻辑
    if feature_to_modify == 'LIMIT_BAL':
        # 信用额度增加或减少 5%
        change_factor = np.random.choice([0.95, 1.05])
        new_customer_df[feature_to_modify] *= change_factor
    elif feature_to_modify == 'AGE':
        # 年龄增加1岁
        new_customer_df[feature_to_modify] += 1
    elif feature_to_modify == 'BILL_AMT1':
        # 账单金额增加或减少 10%
        change_factor = np.random.choice([0.90, 1.10])
        new_customer_df[feature_to_modify] *= change_factor
    elif feature_to_modify == 'PAY_AMT1':
        # 支付金额减少或增加 10% (支付变少可能增加风险)
        change_factor = np.random.choice([0.90, 1.10])
        new_customer_df[feature_to_modify] *= change_factor
    
    # 确保数值不为负
    new_customer_df[new_customer_df < 0] = 0

    
    print(f"尝试修改: {feature_to_modify} -> {new_customer_df[feature_to_modify].iloc[0]:.2f}")
    return new_customer_df


print("--- 准备工作完成！---")
print("\n我们的起点客户画像:")
display(start_customer_df)

# 计算起点和终点的初始embedding和距离
start_embedding = get_embedding_for_df(start_customer_df)
initial_reward = calculate_reward(start_embedding, centroid_high_risk_vip)
print(f"\n起点离'高风险VIP'目标的初始“奖励” (负距离): {initial_reward:.4f}")
```

### 增强版的生成器：

```python

# 增强版的可修改特征列表
modifiable_features = [
    'LIMIT_BAL', 'AGE', 
    'BILL_AMT1', 'PAY_AMT1',
    'PAY_1', 'PAY_2', 'PAY_3' # 我们先加入最近3个月的支付状态
]

def modify_feature(customer_df):
    """(增强版) 随机选择一个特征并进行微小的、合乎逻辑的修改"""
    new_customer_df = customer_df.copy()
    feature_to_modify = np.random.choice(modifiable_features)
    
    print(f"尝试修改: {feature_to_modify}", end=" -> ")
    
    # --- 原有的数值特征修改逻辑 ---
    if feature_to_modify == 'LIMIT_BAL':
        change_factor = np.random.choice([0.95, 1.05])
        new_customer_df[feature_to_modify] *= change_factor
    elif feature_to_modify == 'AGE':
        new_customer_df[feature_to_modify] += 1
    elif feature_to_modify == 'BILL_AMT1':
        change_factor = np.random.choice([0.90, 1.10])
        new_customer_df[feature_to_modify] *= change_factor
    elif feature_to_modify == 'PAY_AMT1':
        change_factor = np.random.choice([0.90, 1.10])
        new_customer_df[feature_to_modify] *= change_factor
    
    # --- 新增的类别特征修改逻辑 ---
    elif feature_to_modify in ['PAY_1', 'PAY_2', 'PAY_3']:
        change = np.random.choice([-1, 1]) # 随机变好或变差
        current_value = new_customer_df[feature_to_modify].iloc[0]
        # 使用np.clip确保值在[-2, 8]的合理范围内
        new_value = np.clip(current_value + change, -2, 8) 
        new_customer_df[feature_to_modify] = new_value
        
    # 确保数值特征不为负 (简单的业务规则)
    for col in ['LIMIT_BAL', 'AGE', 'BILL_AMT1', 'PAY_AMT1']:
         if col in new_customer_df.columns and new_customer_df[col].iloc[0] < 0:
            new_customer_df[col] = 0
    
    print(f"{new_customer_df[feature_to_modify].iloc[0]:.2f}")
    return new_customer_df
```

## 步骤二 循环生成

```python
# --- 第三阶段：步骤二，组装并运行生成循环 ---
import warnings
from pandas.errors import SettingWithCopyWarning

# 忽略特定类型的警告
warnings.simplefilter(action="ignore", category=FutureWarning)
warnings.simplefilter(action="ignore", category=SettingWithCopyWarning)
warnings.simplefilter(action="ignore", category=FutureWarning)
warnings.simplefilter(action="ignore", category=SettingWithCopyWarning)


print("--- 开始定向生成反事实客户画像 ---")

# --- 初始化 ---
current_customer_df = start_customer_df.copy()
current_embedding = start_embedding.copy()
current_reward = initial_reward

# 记录整个“进化”过程
history = []
history.append({
    'step': 0,
    'reward': current_reward,
    'data': current_customer_df.iloc[0].to_dict() # 记录当前客户的数据
})

# --- 设置生成参数 ---
num_steps = 200  # 总共尝试修改多少次
no_improvement_tolerance = 30 # 如果连续这么多次都没有改进，就提前停止
steps_without_improvement = 0

print(f"\n起点客户: \n{start_customer_df[modifiable_features].to_string(index=False)}")
print(f"初始奖励: {current_reward:.4f}")
print("-" * 50)


# --- 生成循环 ---
for i in range(1, num_steps + 1):
    print(f"--- 步骤 {i}/{num_steps} ---")
    
    # 1. 提议一个修改
    proposed_customer_df = modify_feature(current_customer_df)
    
    # 2. 计算新画像的 Embedding
    proposed_embedding = get_embedding_for_df(proposed_customer_df)
    
    # 3. 计算新的奖励
    proposed_reward = calculate_reward(proposed_embedding, centroid_high_risk_vip)
    print(f"新奖励: {proposed_reward:.4f} (当前最佳: {current_reward:.4f})")
    
    # 4. 决定是否接受修改
    if proposed_reward > current_reward:
        print("✅ 接受修改！离目标更近了。")
        current_customer_df = proposed_customer_df
        current_embedding = proposed_embedding
        current_reward = proposed_reward
        steps_without_improvement = 0 # 重置计数器
        
        # 记录成功的步骤
        history.append({
            'step': i,
            'reward': current_reward,
            'data': current_customer_df.iloc[0].to_dict()
        })
    else:
        print("❌ 放弃修改。离目标更远或没有变化。")
        steps_without_improvement += 1

    print("-" * 20)
    
    # 检查是否需要提前停止
    if steps_without_improvement >= no_improvement_tolerance:
        print(f"\n连续 {no_improvement_tolerance} 次没有改进，提前停止生成。")
        break

print("\n--- 生成过程结束 ---")

# --- 结果分析 ---
final_customer_df = current_customer_df
final_reward = current_reward

print(f"初始奖励: {initial_reward:.4f}")
print(f"最终奖励: {final_reward:.4f}")
print(f"奖励提升: {final_reward - initial_reward:.4f}")

print("\n--- 原始客户画像 (部分特征) ---")
print(start_customer_df[modifiable_features].to_string(index=False))

print("\n--- 生成的反事实客户画像 (部分特征) ---")
print(final_customer_df[modifiable_features].to_string(index=False))

# 我们可以将演变历史转换成DataFrame来更方便地查看
history_df = pd.DataFrame([item['data'] for item in history])
history_df['reward'] = [item['reward'] for item in history]
print("\n--- 客户画像演变的关键步骤 ---")
display(history_df[modifiable_features + ['reward']])
```

# 第四阶段 进行模型压力测试

## 步骤一，封装并批量生成反事实样本

```python
# --- 第四阶段：步骤一，封装并批量生成反事实样本 ---

def generate_counterfactual(start_customer_df, target_embedding, max_steps=200):
    """将我们的生成循环封装成一个函数"""
    current_customer_df = start_customer_df.copy()
    start_embedding = get_embedding_for_df(start_customer_df)
    current_embedding = start_embedding.copy()
    current_reward = calculate_reward(current_embedding, target_embedding)
    
    for _ in range(max_steps):
        proposed_customer_df = modify_feature(current_customer_df)
        proposed_embedding = get_embedding_for_df(proposed_customer_df)
        proposed_reward = calculate_reward(proposed_embedding, target_embedding)
        
        if proposed_reward > current_reward:
            current_customer_df = proposed_customer_df
            current_embedding = proposed_embedding
            current_reward = proposed_reward
    
    return current_customer_df

print("--- 开始批量生成100个高风险反事实样本 (这会需要一些时间) ---")

# 从“普通客户”群体中随机选择100个起点
start_customers = df[(df['cluster_id'] == 0) & (df['default'] == 0)].sample(100, random_state=42)

counterfactuals = []
for i in range(len(start_customers)):
    print(f"正在生成第 {i+1}/100 个样本...")
    start_customer = pd.DataFrame(start_customers.iloc[i]).T
    # 目标是高风险/稀疏区域的中心
    cf_customer = generate_counterfactual(start_customer, centroid_high_risk_vip)
    counterfactuals.append(cf_customer)

# 将生成的样本列表合并成一个大的DataFrame
cf_df = pd.concat(counterfactuals, ignore_index=True)

# 给这些生成的“高风险”样本打上“违约”的标签
cf_df['default'] = 1

print("\n--- 100个反事实样本生成完毕！---")
print("生成样本的特征预览：")
display(cf_df[modifiable_features].head())
```

## 步骤二 恢复数据

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, recall_score, accuracy_score
import kagglehub
import os
# --- 0. 重新加载和划分数据以恢复 test_df ---
print("--- 步骤0: 正在恢复 train_df, val_df, test_df ---")

# a. 加载完整数据集
dataset_dir = kagglehub.dataset_download("uciml/default-of-credit-card-clients-dataset")
dataset_path = os.path.join(dataset_dir,"UCI_Credit_Card.csv")
df_full = pd.read_csv(dataset_path) # 使用新变量名 df_full 以免混淆
df_full = df_full.drop("ID",axis = 1)
df_full = df_full.rename(columns={'default.payment.next.month': 'default', 'PAY_0': 'PAY_1'})

# b. 定义特征类型 (确保和训练时完全一致)
categorical_cols = ['SEX', 'EDUCATION', 'MARRIAGE'] + [f'PAY_{i}' for i in range(1, 7)]
target_col = 'default'
for col in categorical_cols:
    df_full[col] = df_full[col].astype('category')

# c. 执行和训练前完全相同的划分步骤
# 使用 random_state=42 保证每次划分结果都一样
train_val_df, test_df = train_test_split(df_full, test_size=0.2, random_state=42, stratify=df_full[target_col])
train_df, val_df = train_test_split(train_val_df, test_size=0.1, random_state=42, stratify=train_val_df[target_col])

print(f"test_df 恢复成功，包含 {len(test_df)} 条样本。")
```

## 步骤三，进行压力测试

```python
from sklearn.metrics import roc_auc_score, recall_score, accuracy_score

print("--- 开始进行模型压力测试 ---")

# 原始测试集
original_test_df = test_df # 假设 test_df 还在内存中

# 创建增强测试集
enhanced_test_df = pd.concat([original_test_df, cf_df], ignore_index=True)

def evaluate_performance(test_set, set_name):
    print(f"\n--- 在【{set_name}】上评估性能 ---")
    
    # 获取预测结果
    preds = loaded_model.predict(test_set)
    
    # (调试用的打印语句可以保留，也可以删掉了，因为我们已经找到问题了)
    # print("模型预测结果 (preds) 的列名:", preds.columns)
    # display(preds.head())
    
    true_labels = test_set['default']
    pred_labels = preds['default_prediction']
    
    # !! 关键修改 !!
    # 将错误的列名 '1_probability' 替换成我们发现的正确列名 'default_1_probability'
    probability_column_name = 'default_1_probability'
    pred_probs = preds[probability_column_name]
    
    # 后续的计算现在可以正常工作了
    acc = accuracy_score(true_labels, pred_labels)
    auc = roc_auc_score(true_labels, pred_probs)
    recall = recall_score(true_labels, pred_labels)
    
    print(f"准确率 (Accuracy): {acc:.4f}")
    print(f"AUC: {auc:.4f}")
    print(f"召回率 (Recall): {recall:.4f}")
    return {'acc': acc, 'auc': auc, 'recall': recall}

# 在原始测试集上评估
original_metrics = evaluate_performance(original_test_df, "原始测试集")

# 在增强测试集上评估
enhanced_metrics = evaluate_performance(enhanced_test_df, "增强测试集 (含反事实样本)")

print("\n--- 压力测试总结 ---")
print(f"加入反事实样本后，模型的AUC从 {original_metrics['auc']:.4f} 变化为 {enhanced_metrics['auc']:.4f}")
print(f"召回率 (识别真实风险的能力) 从 {original_metrics['recall']:.4f} 变化为 {enhanced_metrics['recall']:.4f}")
```
